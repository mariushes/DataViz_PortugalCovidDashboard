```python
def date_transform(date):
    date_list = date.split("/")
    day = date_list[1]
    if len(day)==1:
        day = "0"+day
    month = date_list[0]
    if len(month)==1:
        month = "0"+month
    year = date_list[2]
    new_string = "20"+year+"-"+month+"-"+day
    return new_string
date_transform("11/12/20")
```

```python
df = pd.read_csv("../Data/time_series_covid19_confirmed_global.csv")
columns = df.columns.to_list()
for i, column in enumerate(columns):
    if i >= 4:
        columns[i] = date_transform(column)
df.columns = columns
df_confirmed_Portugal = df[df["Country/Region"]=="Portugal"]
df_confirmed_Portugal.to_csv("../Data/time_series_covid19_confirmed_portugal.csv", index=False)
```

## Daily new cases per 100k

```python
australian_new_infections_per100k = df_confirmed_new_Australia.copy()
australian_new_infections_per100k = australian_new_infections_per100k.rename(columns={'Province/State': 'State'})
australian_new_infections_per100k = australian_new_infections_per100k.set_index('State')
australian_new_infections_per100k
```

```python
state_population = pd.read_csv('../Data/Raw/australian_states_population.csv', index_col='State')
state_population
```

```python
australian_new_infections_per100k = state_population.join(australian_new_infections_per100k)
australian_new_infections_per100k.head(10)
```

```python
timeline_columns = australian_new_infections_per100k.columns[5:]
def apply_perCapita(row):
    row[timeline_columns] = row[timeline_columns].apply(lambda x: x/(row.Population/100000))
    return row
```

```python
australian_new_infections_per100k = australian_new_infections_per100k.apply(apply_perCapita, axis=1)
australian_new_infections_per100k = australian_new_infections_per100k.reset_index()
australian_new_infections_per100k.to_csv("../Data/Processed/time_series_covid19_confirmed_new_infections_per100k_australia.csv", index=False)
australian_new_infections_per100k
```

## Seven Day Average 

```python
def seven_day_average(df):
    df_new = df.copy()
    for i in range(len(df_new.index)):
        for j in range(1,len(df_new.columns)):
            if j >=10:
                df_new.iloc[i,j] = df.iloc[i,j-6:j].sum() / 7
    return df_new
```

```python
pd.set_option('display.max_columns', 30)
df_confirmed_new_Australia_7average = seven_day_average(df_confirmed_new_Australia)
```

```python
df_confirmed_new_Australia_7average.to_csv("../Data/Processed/time_series_covid19_confirmed_new_infections_7average_australia.csv", index=False)
```

```python
df_confirmed_new_Portugal_7average = seven_day_average(df_confirmed_new_Portugal)
df_confirmed_new_Portugal_7average.to_csv("../Data/time_series_covid19_confirmed_new_infections_7average_portugal.csv", index=False)
```

```python
def transform_to_date_index(df):
    df_transposed = df.transpose()
    df_transposed = df_transposed.drop(axis=0, labels=["Country/Region", "Lat","Long"])
    new_columns = df.iloc[:,0].to_list()
    df_transposed.columns=new_columns
    df_transposed = df_transposed.drop(axis=0, labels=["Province/State"])
    new_indexs = df_transposed.index.to_list()
    return df_transposed
```

```python
def date_transform(date):
    date_list = date.split("/")
    day = date_list[1]
    if len(day)==1:
        day = "0"+day
    month = date_list[0]
    if len(month)==1:
        month = "0"+month
    year = date_list[2]
    new_string = "20"+year+"-"+month+"-"+day
    return new_string
date_transform("11/12/20")
```

```python
df_confirmed_new_Australia_7average_transformed = transform_to_date_index(df_confirmed_new_Australia_7average)
display(df_confirmed_new_Australia_7average_transformed)
```

```python
df_testing_australia_date_index = df_testing_australia.copy()
#df_testing_australia_date_index.index = df_testing_australia_date_index["Date"]
df_testing_australia_date_index = df_testing_australia_date_index.set_index("Date")
df_join_testing_cases = df_testing_australia_date_index.join(df_confirmed_new_Australia_7average_transformed)
df_join_testing_cases
```

```python
df_population = pd.read_csv("../Data/Raw/australian_states_population.csv")
australia_population = df_population.iloc[0,2]
df_population
```

```python
positive_rates = []
sum_tests = 0
for i in range(len(df_join_testing_cases.index)):
    tests_per_thousand = df_join_testing_cases.iloc[i,1]
    tests = tests_per_thousand/1000 * australia_population
    sum_tests += tests
    new_confirmed_cases = df_join_testing_cases.iloc[i,10]
    positive_rate = new_confirmed_cases/tests
    positive_rates.append(positive_rate)
print(positive_rates)
```

```python
df_testing_positive_rate = df_join_testing_cases.copy()
df_testing_positive_rate["Positive Rate"] = positive_rates
df_testing_positive_rate=df_testing_positive_rate.drop(axis=1, labels=["Australia","New South Wales","Northern Territory","Queensland","South Australia","Tasmania","Victoria","Western Australia"])
dates=df_testing_positive_rate.index
df_testing_positive_rate=df_testing_positive_rate.reset_index()
df_testing_positive_rate["Date"] =dates
df_testing_positive_rate
```
