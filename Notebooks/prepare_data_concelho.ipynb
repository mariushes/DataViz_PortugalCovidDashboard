```python
def transform_date_string(date):
    splits = date.split("-")
    return splits[2]+"-"+splits[1]+"-"+splits[0]
transform_date_string("24-03-2020")
```

```python
new_data_dict = dict()
risk_dict = dict()
dates = []
concelhos = []
for i in range(len(df2.index)):
    row = df2.iloc[i].tolist()
    date = transform_date_string(row[0])
    concelho = row[1]
    new_infections = row[3]
    risk_level = row[6]
    if risk_level =="Moderada" or risk_level =="Baixo a Moderado" or risk_level =="Moderado" or risk_level == "Baixa":
        risk_level = 1
    elif risk_level =="Elevada" or risk_level =="Elevado":
        risk_level = 2
    elif risk_level == "Muito elevada" or risk_level == 'Muito Elevado':
        risk_level = 3
    elif risk_level == "Extremamente Elevado":
        risk_level = 4
    elif risk_level == "Baixa" or risk_level == "Muito baixa":
        risk_level = 0
    if concelho not in concelhos:
        concelhos.append(concelho)
    if date not in dates:
        new_data_dict[date] = dict()
        risk_dict[date] = dict()
        dates.append(date)
    new_data_dict[date][concelho]= new_infections
    risk_dict[date][concelho] = risk_level
```

```python
def date_range(date1, date2):
    date1 = datetime.date.fromisoformat(date1)
    date2 = datetime.date.fromisoformat(date2)
    dates = []
    for n in range(int ((date2 - date1).days)+1):
        dates.append((date1 + datetime.timedelta(n)).isoformat())
    return dates
```
