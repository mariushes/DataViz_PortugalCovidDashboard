```python
new_data_dict = dict()
dates = []
concelhos = []
for i in range(len(df2.index)):
    row = df2.iloc[i].tolist()
    date = row[0]
    concelho = row[1]
    new_infections = row[3]
    if concelho not in concelhos:
        concelhos.append(concelho)
    if date not in dates:
        new_data_dict[date] = dict()
        dates.append(date)
    new_data_dict[date][concelho]= new_infections
```

```python
def transform_date_string(date):
    splits = date.split("-")
    return splits[2]+"-"+splits[1]+"-"+splits[0]
transform_date_string("24-03-2020")
```

```python
df1
```

```python
df_merged = df1.copy()
columns = df_merged.columns.tolist()
columns.pop(0)
rows = []
for date in dates:
    row = []
    row.append(date)
    for concelho in columns:
        row.append(new_data_dict[date][concelho])
    rows.append(row)
df_new = pd.DataFrame(rows, columns=df_merged.columns)
df_merged = df_merged.append(df_new)
new_columns = [x.lower() for x in df_merged.columns.tolist()]
for i, x in enumerate(new_columns):
    if x == "ponte de sor":
        new_columns[i]='ponte de sôr'
df_merged.columns = new_columns
df_merged.loc[:,"data"]= [transform_date_string(x) for x in df_merged.loc[:,"data"].tolist()]
df_merged=df_merged.fillna(0)
```

```python


df_new_per100k = df_merged.copy()
sums = []
sums_sum = 0
for column in df_new_per100k.columns:
    if column == "data":
        continue
    #population = df_population.at[column,"population"]
    summed = df_new_per100k[column].sum()
    sums.append(summed)
    sums_sum += summed
sums_sum
```

```python
df_population = df2.copy()
df_population = df_population[["data", "concelho", "population"]]
df_population = df_population[df_population["data"]=="11-11-2020"]
df_population = df_population[["concelho", "population"]]
df_population["concelho"] = [x.lower() for x in df_population["concelho"].tolist()]
df_population = df_population.replace(to_replace="ponte de sor",value='ponte de sôr')
df_population = df_population.set_index("concelho")
df_population
```

```python
df_new_per100k = df_merged.copy()
for column in df_new_per100k.columns:
    if column == "data":
        continue
    population = df_population.at[column,"population"]
    for i in df_new_per100k.index:
        #print(i, column)
        #print(df_new_per100k.at[i,column])
        df_new_per100k.at[i,column] = (df_new_per100k.at[i,column]/population)*100000
```
