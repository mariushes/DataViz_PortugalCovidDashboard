```python
def color_interval(df,start, end, color):
    date_list = date_range(start, end)
    values = df.loc[0,date_list]
    nan_array = np.empty(len(date_list))
    nan_array[:] = np.NaN
    #df.loc[0,date_list] = nan_array
    nan_array = np.empty(len(df.columns))
    nan_array[:] = np.NaN
    new_position = len(df.index)
    df.loc[new_position] = nan_array
    df.loc[new_position,date_list] = values
    df.loc[new_position,"color"] = color
    return df
```

```python
def date_range(date1, date2):
    date1 = datetime.date.fromisoformat(date1)
    date2 = datetime.date.fromisoformat(date2)
    dates = []
    for n in range(int ((date2 - date1).days)+1):
        dates.append((date1 + datetime.timedelta(n)).isoformat())
    return dates
```

```python
portugal_new_infections_7average_data_color = portugal_new_infections_7average_data.copy()
portugal_new_infections_7average_data_color["color"] = ["green"]
color_interval(portugal_new_infections_7average_data_color, "2020-01-27", "2020-01-29", "white")
```

```python
slider_date = "2021-02-01"
figure = go.Figure(data=data,
                        layout=layout)
figure.add_vline(x=slider_date)
figure.show()
```

```python
import re

pattern = re.compile('(\w+\s)+')
def extract_state(text):
    match = pattern.search(text)
    if(match):
        span = match.span()
        return text[span[0]: span[1]-1]
    
    return ''
```

```python
#australian_state_perCapita['Province/State'] = australian_state_perCapita['Province/State'].apply(extract_state)
australian_state_perCapita = australian_state_perCapita.rename(columns={'Province/State': 'State'})
australian_state_perCapita = australian_state_perCapita.set_index('State')
australian_state_perCapita
```

```python
timeline_columns = australian_state_perCapita.columns[2:]
def apply_perCapita(row):
    row[timeline_columns] = row[timeline_columns].apply(lambda x: x/(row.Population/100000))
    return row
```

```python
australian_state_perCapita = australian_state_perCapita.apply(apply_perCapita, axis=1)
australian_state_perCapita = australian_state_perCapita.reset_index()
australian_state_perCapita.head(10)
```

```python
def date1_greater_date2(date1, date2):
    date1_list = date1.split("-")
    year1 = date1_list[0]
    month1 = date1_list[1]
    day1 = date1_list[2]
    date2_list = date2.split("-")
    year2 = date2_list[0]
    month2 = date2_list[1]
    day2 = date2_list[2]
    if year1 > year2: return True
    elif year1 < year2: return False
    elif year1 == year2:
            if month1 > month2: return True
            elif month1 < month2: return False
            elif month1 == month2:
                if day1 > day2: return True
                else: return False
```

```python
def remove_data_timespan(df, start, end, state):
    df = df.copy()
    state_index = 999
    for i in range(len(df["State"])):
        if df["State"].tolist()[i] == state:
            state_index = i
    for i in range(len(df.columns)):
        if i >= 3:
            date = df.columns[i]
            if date1_greater_date2(date, start) and date1_greater_date2(end, date) :
                df.iloc[state_index,i] = None
    return df
```
